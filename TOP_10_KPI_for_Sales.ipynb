# Essential Sales KPIs by python, Every Data Scientist Should Know thumbnail_1.jpeg

Sales KPIs are measurable metrics that indicate how well a company meets its sales objectives and goals. These sales metrics assist in evaluating the effectiveness of sales teams, departments, or people by offering insights into many areas of the sales process.

I explore how these essential sales KPIs can provide valuable insights for making informed business decisions.

**1. Sales Growth**
Measures the increase in sales over a specific period.

**Formula:**
Sales Growth
=
(
Current Period Sales
−
Previous Period Sales
)/
Previous Period Sales
×
100



```python
import pandas as pd
import matplotlib.pyplot as plt
# Sample data
data = {'Month': ['January', 'February', 'March', 'April'],
        'Sales': [2000, 2500, 3000, 4000]}
df = pd.DataFrame(data)

# Calculate percentage change in Sales
df['Sales Growth'] = df['Sales'].pct_change() * 100

print(df)
```

```python
# Visualization
plt.plot(df['Month'], df['Sales Growth'], marker='o')
plt.title('Sales Growth Over Time')
plt.xlabel('Month')
plt.ylabel('Sales Growth (%)')
plt.show()
```

**2. Sales Target**
The sales goal set for a specific period.

**Formula:**
Sales Target Achievement
=
Actual Sales/
Target Sales
×
100
S


```python
# Sample data
df['Target Sales'] = [2200, 2700, 3200, 3500]

# Calculate Sales Target Achievement
df['Target Achievement'] = (df['Sales'] / df['Target Sales']) * 100

# Visualization
plt.plot(df['Month'], df['Target Achievement'], marker='o')
plt.title('Sales Target Achievement')
plt.xlabel('Month')
plt.ylabel('Achievement (%)')
plt.show()
```

**3. Conversion Rate**
The percentage of prospects who become customers.

**Formula:**
Conversion Rate
=
Number of Sales/
Number of Samples
×
100


```python
# Sample data
data = {'Month': ['January', 'February', 'March', 'April'],
        'Samples': [100, 150, 200, 250],
        'Sales': [20, 30, 50, 60]}
df = pd.DataFrame(data)

# Calculate Conversion Rate
df['Conversion Rate'] = (df['Sales'] / df['Samples']) * 100

# Visualization
plt.plot(df['Month'], df['Conversion Rate'], marker='o')
plt.title('Conversion Rate Over Time')
plt.xlabel('Month')
plt.ylabel('Conversion Rate (%)')
plt.show()
```

**4-Customer Acquisition Cost (CAC)**
The cost associated with acquiring a new customer.

**Formula:**
CAC
=
Total Sales and Marketing Expenses/
Number of New Customers Acquired


```python
# Sample data
data = {'Month': ['January', 'February', 'March', 'April'],
        'Marketing Expenses': [1000, 1200, 1500, 1800],
        'New Customers': [10, 15, 20, 25]}
df = pd.DataFrame(data)

# Calculate CAC
df['CAC'] = df['Marketing Expenses'] / df['New Customers']

# Visualization
plt.plot(df['Month'], df['CAC'], marker='o')
plt.title('Customer Acquisition Cost')
plt.xlabel('Month')
plt.ylabel('CAC ($)')
plt.show()
```

**5- Churn Rate**
The percentage of customers lost over a specific period.

**Formula:**
Churn Rate
=
Number of Customers Lost
Total Number of Customers at the Start of Period
×
100


```python
# Sample data
data = {'Month': ['January', 'February', 'March', 'April'],
        'Customers at Start': [100, 120, 150, 180],
        'Customers Lost': [5, 10, 8, 15]}
df = pd.DataFrame(data)

# Calculate Churn Rate
df['Churn Rate'] = (df['Customers Lost'] / df['Customers at Start']) * 100

# Visualization
plt.plot(df['Month'], df['Churn Rate'], marker='o')
plt.title('Customer Churn Rate')
plt.xlabel('Month')
plt.ylabel('Churn Rate (%)')
plt.show()
```

**6-Average Customer Lifespan**
 is the average duration (typically in years) that a customer continues to purchase from a company before churning.

 Average Customer Lifespan= 1/ Customer Churn Rate



```python
import pandas as pd

# Sample data: Number of customers at the start and lost during the period
data = {'Month': ['January', 'February', 'March', 'April'],
        'Customers at Start': [100, 120, 150, 180],
        'Customers Lost': [5, 10, 8, 15]}
df = pd.DataFrame(data)

# Calculate Churn Rate
df['Churn Rate'] = df['Customers Lost'] / df['Customers at Start']

# Calculate Average Customer Lifespan for each month
df['Average Customer Lifespan'] = 1 / df['Churn Rate']

print(df[['Month', 'Churn Rate', 'Average Customer Lifespan']])
```

**7. Customer Lifetime Value (CLV)**
Definition: The total revenue expected from a customer over their lifetime.

**Formula:**
CLV
=
Average Purchase Value
×
Number of Purchases per Year
×
Average Customer Lifespan


```python
# Sample data
data = {'Customer': ['A', 'B', 'C', 'D'],
        'Average Purchase Value': [100, 150, 200, 250],
        'Purchases per Year': [10, 8, 12, 5],
        'Average Customer Lifespan': [5, 6, 4, 3]}
df = pd.DataFrame(data)

# Calculate CLV
df['CLV'] = df['Average Purchase Value'] * df['Purchases per Year'] * df['Average Customer Lifespan']

# Visualization
plt.bar(df['Customer'], df['CLV'])
plt.title('Customer Lifetime Value')
plt.xlabel('Customer')
plt.ylabel('CLV ($)')
plt.show()
```

**8. Customer Retention Rate** Measures the percentage of customers a company retains over a specific period.

**Formula:**
Customer Retention Rate
=
(
Customers at End of Period
−
New Customers Acquired
Customers at Start of Period
)
×
100


```python
# Sample data
data = {'Month': ['January', 'February', 'March', 'April'],
        'Customers at Start': [100, 120, 150, 180],
        'Customers at End': [120, 150, 180, 200],
        'New Customers Acquired': [30, 40, 50, 60]}
df = pd.DataFrame(data)

# Calculate Customer Retention Rate
df['Customer Retention Rate'] = ((df['Customers at End'] - df['New Customers Acquired']) / df['Customers at Start']) * 100

# Visualization
plt.plot(df['Month'], df['Customer Retention Rate'], marker='o')
plt.title('Customer Retention Rate')
plt.xlabel('Month')
plt.ylabel('Retention Rate (%)')
plt.show()
```

**9. Customer Acquisition Efficiency**
**Definition:** Measures how efficiently a company acquires new customers in terms of revenue generated per dollar spent on acquisition.

**Formula:**
Customer Acquisition Efficiency
=
Total Revenue/Marketing Expenses



```python
# Sample data
data = {'Month': ['January', 'February', 'March', 'April'],
        'Total Revenue': [200000, 250000, 300000, 400000],
        'Marketing Expenses': [100000, 120000, 150000, 180000]}
df = pd.DataFrame(data)

# Calculate Customer Acquisition Efficiency
df['Customer Acquisition Efficiency'] = df['Total Revenue'] / df['Marketing Expenses']

# Visualization
plt.plot(df['Month'], df['Customer Acquisition Efficiency'], marker='o')
plt.title('Customer Acquisition Efficiency')
plt.xlabel('Month')
plt.ylabel('Revenue per Dollar Spent')
plt.show()
```


**10. Repeat Purchase Rate**
Definition: Measures the percentage of customers who make more than one purchase over a specific period.

Formula:
Repeat Purchase Rate
=
(
Number of Repeat Customers/
Total Number of Customers
)
×
100


```python
# Sample data
data = {'Month': ['January', 'February', 'March', 'April'],
        'Total Customers': [100, 120, 150, 180],
        'Repeat Customers': [30, 40, 60, 80]}
df = pd.DataFrame(data)

# Calculate Repeat Purchase Rate
df['Repeat Purchase Rate'] = (df['Repeat Customers'] / df['Total Customers']) * 100

# Visualization
plt.plot(df['Month'], df['Repeat Purchase Rate'], marker='o')
plt.title('Repeat Purchase Rate')
plt.xlabel('Month')
plt.ylabel('Repeat Purchase Rate (%)')
plt.show()
```
