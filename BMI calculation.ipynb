
## Python script for BMI calculation in ColorfulAnalysis

**This script does the following:**

- **Takes user input for weight and height.**
  - Uses the `input` function to prompt the user for weight and height in kilograms and meters.
  - Converts the input to `float` for numeric calculations.

- **Calculates BMI using the provided formula.**
  - Utilizes the BMI formula: (BMI = {{weight}\{height}^2}}).

- **Categorizes the BMI into "Underweight," "Normal Weight," "Overweight," or "Obese."**
  - Defines categories based on standard BMI ranges.
  - Uses conditional statements to determine the category.

- **Displays the calculated BMI and its category.**
  - Prints the calculated BMI with two decimal places.
  - Prints the determined BMI category.

- **Visualizes the BMI category using a bar chart.**
  - Utilizes the `matplotlib` library for data visualization.
  - Creates a bar chart to represent the distribution of BMI categories.

This script is designed to be interactive, providing users with an understanding of their BMI category through user input, calculation, categorization, and visualization.

```python
import matplotlib.pyplot as plt

# Function to calculate BMI
def calculate_bmi(weight, height):
    height_m =height/100
    bmi = weight / (height_m ** 2)
    return bmi

# Function to categorize BMI
def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal Weight"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Get user input for weight and height
weight = float(input("Enter your weight in KG: "))
height = float(input("Enter your height in centimeters: "))

# Calculate BMI
bmi_result = calculate_bmi(weight, height)

# Categorize BMI
bmi_category = categorize_bmi(bmi_result)

# Display results
print(f"\nYour BMI is {bmi_result:.2f}")
print(f"You are categorized as {bmi_category}")

# Simple bar chart for BMI
categories = ["Underweight", "Normal Weight", "Overweight", "Obese"]
bmi_values = [0, 0, 0, 0]

category_index = categories.index(bmi_category)
bmi_values[category_index] += 1

# Plotting the bar chart
plt.bar(categories, bmi_values, color=['blue', 'green', 'yellow', 'red'])
plt.title("BMI Category Distribution")
plt.xlabel("BMI Category")
plt.ylabel("Number of Individuals")
plt.show()
```
